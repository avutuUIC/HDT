```python
prompt = f"""
You are a virtual coach that generates personalized physical training programs. Consider the following digital twin data for Samantha Cox.

Personal Profile:
- Age: 35
- Gender: Female
- Height: 179 cm
- Weight: 99.3 kg
- Body Fat %: 24.9
- Fitness Level: Beginner
- Preferred Training Types: cardio, yoga, strength, mobility
- Recent Activities: April 7 – Cycling (76 min, 363 kcal), April 6 – Running (77 min, 590 kcal)
- Injury History: Hamstring Pull (not fully recovered)
- Environmental Conditions: 27.8°C, 70% humidity, Air Quality Index 66

Mental State:
- Stress Level: 7
- Anxiety Level: High
- Motivation Score: 4

Physical State:
- Energy Level: Normal
- Fatigue Level: Medium
- Strength Score: 3.1
- Endurance Score: 5.2
- Recovery Index: 4.4
- Injury Risk: 0.62

Cognitive State:
- Focus Level: Normal
- Decision-Making: Average
- Reaction Time: 390 ms
- Accuracy: 0.84

Biomechanical State:
- Balance Score: 4.7
- Dexterity Score: 5.0
- Posture: Slouched

Interpersonal State:
- Leadership Score: 3.2
- Social Engagement: Normal
- Cooperation Score: 7.6

Based on all this information, generate a safe and effective 1-day training regime for Samantha that:
- Accommodates her hamstring injury
- Enhances motivation and social engagement
- Balances intensity based on current physical and cognitive readiness
- Incorporates recovery and posture improvement
- Uses her preferences (cardio, yoga, etc.)

Format the response clearly with sections for Warm-Up, Main Workout, and Cool-Down, including estimated duration and notes.
"""
```

```python
import openai

openai.api_key = ".."

response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=[
        {"role": "system", "content": "You are a helpful and knowledgeable fitness coach."},
        {"role": "user", "content": prompt}
    ],
    temperature=0.7,
    max_tokens=800
)

print(response['choices'][0]['message']['content'])
```
